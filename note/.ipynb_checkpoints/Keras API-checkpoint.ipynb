```python
import tensorflow as tf
import time
from tensorflow.keras import layers
```

# 创建网络

```python
inputs = tf.keras.Input(shape=(784,), name='in_img')
h1 = layers.Dense(32, activation='relu')(inputs)
h2 = layers.Dense(32, activation='relu')(h1)
outputs = layers.Dense(10, activation='softmax')(h2)
model = tf.keras.Model(inputs=inputs, outputs=outputs, name='mnist_model')
```

```python
model.summary
```

```python
tf.keras.utils.plot_model(model, '/Users/zqy/Downloads/mnist_model.png', show_shapes=True)
# see https://blog.csdn.net/wushaowu2014/article/details/88958400?utm_medium=distribute.pc_relevant.none-task-blog-OPENSEARCH-1.control&depth_1-utm_source=distribute.pc_relevant.none-task-blog-OPENSEARCH-1.control
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train = x_train.reshape(x_train.shape[0], 784).astype('float32') / 255
x_test = x_test.reshape(x_test.shape[0], 784).astype('float32') / 255
```

```python
model.compile(optimizer=tf.keras.optimizers.RMSprop(),
             loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
             )
```

```python
history = model.fit(x_train, y_train, batch_size=64, epochs=5, validation_split=0.2)
```

```python
test_scores = model.evaluate(x_test, y_test, verbose=0)
print('test loss:', test_scores[0])
print('test acc:', test_scores[1])
```
